            double side1 = Convert.ToDouble(Console.ReadLine());

        // the method takes 3 sides of a triangle as input and return the area of the triangle as a double datatype  
        public double triangleArea(double a, double b, double c)  
        {  
            //Calculating the semiperimeter  
            double semiperimeter = (a + b + c) / 2;  
            //using Math.sqrt class for finding the square root and returning the area using Herons Formula  
            return Math.Sqrt(semiperimeter * ((semiperimeter - a) * (semiperimeter - b) * (semiperimeter - c)));  
        }  
    }

# Question 6  
You can turn a word into pig-Latin using the following two rules (simplified):  
 If the word starts with a consonant, move that letter to the end and append   
'ay'. For example, 'happy' becomes 'appyhay' and 'pencil' becomes 'encilpay'.  
 If the word starts with a vowel, simply append 'way' to the end of the word.   
For example, 'enter' becomes 'enterway' and 'other' becomes 'otherway' . For   
our purposes, there are 5 vowels: a, e, i, o, u (so we count y as a consonant).  
Write a function pig() that takes a word (i.e., a string) as input and returns its pigLatin form. Your function should still  work if the input word contains upper case   
characters. Your output should always be lower case however.   
'>>> pig('happy')  
'appyhay'  
'>>> pig('Enter')  
'enterway'  

# Answer 6

```python
# Taking a word as input and storing it in a variable
word=input("Enter the desired word: ")

# Converting to lower case so that the comparison of the vowels succeed irrespective of the Case (upper /lower)
s=word.lower()  

# Defining the method and first checking the first charcter of the word and the manipulating the word accordingly using string methods
def pig(s):  
    if ((s[0]=='a')or(s[0]=='e')or(s[0]=='i')or(s[0]=='o')or(s[0]=='u')):  
        out=s+"way"  
         
    else:  
        out=s[1:]+s[0]+"ay"  
         

    return out
     
print(pig(s))
```

# Question 7  
File bloodtype1.txt records blood-types of patients (A, B, AB, O or OO) at a clinic.   
Write a function bldcount() that reads the file with name name and reports (i.e.,   
prints) how many patients there are in each bloodtype.  
'>>> bldcount('bloodtype.txt')  
There are 10 patients of blood type A.  
There is one patient of blood type B.  
There are 10 patients of blood type AB.  
There are 12 patients of blood type O.  
There are no patients of blood type OO.  


# Answer 7

```python
#Reading the file using absolute path and in read mode  
infile = open("bloodtype1.txt",'r')  
data = infile.read()  
#storing all the different occurences in array as there is a space in between thus using same for splitting  
words = data.split()  
#initializing variables to store the count of each Blood type  
countA = 0  
countB = 0  
countAB = 0  
countO = 0  
countOO = 0  
#using loop to iterate over array having all occurences and checking the type and increasing the count  
for i in words:  
    if i=='A':  
        countA+=1  
    elif i=='B':  
        countB+=1  
    elif i=='AB':  
        countAB+=1  
    elif i=='O':  
        countO+=1  
    elif i=='OO':  
        countOO+=1  
#pring the count for all the blood type  
print("Total People with blood type A : "+str(countA))  
print("Total People with blood type B : "+str(countB))   
print("Total People with blood type AB : "+str(countAB))  
print("Total People with blood type O : "+str(countO))  
print("Total People with blood type OO : "+str(countOO))  

```
